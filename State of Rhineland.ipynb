```python
import pandas as pd

# Define desired funding type order
funding_order = ['Pre-Seed', 'Seed', 'Series A', 'Series B']

# Aggregation function for each year
def aggregate_yearly_data(df, year):
    """
    Aggregates total funding rounds and investment sums per funding type for a given year.
    """
    total_rounds = len(df)
    total_investment = df['Money Raised'].sum()

    # Count rounds per funding type
    rounds_per_type = df['Funding Type'].value_counts().reindex(funding_order, fill_value=0)

    # Sum investment per funding type
    investment_per_type = df.groupby('Funding Type')['Money Raised'].sum().reindex(funding_order, fill_value=0)

    # Build summary dictionary
    data = {
        'Year': year,
        'Total Rounds': total_rounds,
        'Total Investment ($M)': total_investment / 1e6,  # Convert to millions
    }

    for ft in funding_order:
        data[f'{ft} Rounds'] = rounds_per_type[ft]
        data[f'{ft} Investment ($M)'] = investment_per_type[ft] / 1e6  # Convert to millions

    return data

# List of DataFrames for each year (ensure these are loaded)
dfs = {
    2021: df_2021,
    2022: df_2022,
    2023: df_2023,
    2024: df_2024,
    2025: df_2025
}

# Create summary DataFrame
summary_df = pd.DataFrame([aggregate_yearly_data(df, year) for year, df in dfs.items()])

summary_df
```

```python
# Define categories
categories = ['Life Science', 'Hardware', 'Software']

# Initialize a dictionary to store the top investor(s) per category
top_investors_by_size = {}

for category in categories:
    # Filter DataFrame by category
    df_category = df_2024[df_2024['Category'] == category]

    # Explode investor names to assign investment amounts per investor
    investors_exploded = (
        df_category[['Investor Names', 'Money Raised']]
        .dropna()
        .assign(Investor=df_category['Investor Names'].str.split(','))
        .explode('Investor')
        .assign(Investor=lambda x: x['Investor'].str.strip())  # Clean whitespace
    )

    # Sum total investment amount per investor
    top_investors = (
        investors_exploded.groupby('Investor')['Money Raised']
        .sum()
        .sort_values(ascending=False)
    )

    # Check for distinct top investor
    if not top_investors.empty:
        max_investment = top_investors.iloc[0]  # Get the highest investment amount
        top_investor_list = top_investors[top_investors == max_investment].index.tolist()

        # Store results
        if len(top_investor_list) == 1:
            top_investors_by_size[category] = top_investor_list[0]
        else:
            top_investors_by_size[category] = top_investor_list  # Store all tied investors
    else:
        top_investors_by_size[category] = "No investors found"

# Display the results
print("Most active investors per category by investments sums")
for category, investor in top_investors_by_size.items():
    if isinstance(investor, list):
        print(f"{category}: Top Investors (Tied) - {', '.join(investor)}")
    else:
        print(f"{category}: Top Investor - {investor}")
```
